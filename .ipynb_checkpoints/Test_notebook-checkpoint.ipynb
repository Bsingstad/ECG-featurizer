```python
def aggregate_diagnostic(y_dic):
    tmp = []
    for key in y_dic.keys():
        if key in agg_df.index:
            tmp.append(agg_df.loc[key].diagnostic_class)
    return list(set(tmp))

demo_data['diagnostic_superclass'] = demo_data.scp_codes.apply(aggregate_diagnostic)
```

```python
manual_annot = np.zeros((9,len(unannotated_iter)))
for n_annot, unannotated in enumerate(unannotated_iter):
    temp_r = 0
    r_counter = 0
    ecgdata, fs, leads, sig_len = _read_directory_data(unannotated,directory)
    signal_length = np.arange(sig_len)
    try:
        _, rpeaks = nk.ecg_peaks(ecg_data[1], sampling_rate=fs)
    except:
        print("Could not find R-peaks")
        fig = go.Figure()
        fig.add_trace(go.Scatter(x=signal_length, y=ecgdata[1],
                mode='lines',
                name='lines'))
        fig.show()
        print("Please help annotate")
        rpeaks = []
        while temp_r != "ok":
            r_counter = r_counter + 1 
            print("Annotate peak. If previous peak was the last peak type ok")
            temp_r = input("Annotate {} R-peak: ".format(r_counter))
            rpeaks.append(temp_r)
        rpeaks = np.asarray(rpeaks)
        rpeaks=rpeaks[0:-1].astype(float)
    
    try:
        _, waves_dwt = nk.ecg_delineate(ecg_data[1], rpeaks, sampling_rate=fs, method="dwt") 
    except:
        print("Could not find do discrete wavelet transformation")
        fig = go.Figure()
        fig.add_trace(go.Scatter(x=signal_length, y=ecgdata[1],
            mode='lines',
            name='lines'))
        fig.show()
        print("Please help annotate")
        p_offsets = []
        p_onsets = []
        p_peaks = []
        t_peaks = []
        t_offsets = []
        t_onsets = []
        for i in range(6):
            temp_dwt = 0
            dwt_counter = 0
            while temp_dwt != "ok":
                dwt_counter = dwt_counter + 1 
                if i == 0:
                    print("Annotate P-offsets. If previous peak was the last P-offset type ok")
                    temp_dwt = input("Annotate {} P-offsets: ".format(dwt_counter))
                    p_offsets.append(temp_dwt)
                elif i == 1:
                    print("Annotate P-onsets. If previous peak was the last P-onsets type ok")
                    temp_dwt = input("Annotate {} P-onsets: ".format(dwt_counter))
                    p_onsets.append(temp_dwt)
                elif i == 2:
                    print("Annotate P-peaks. If previous peak was the last P-peaks type ok")
                    temp_dwt = input("Annotate {} P-peaks: ".format(dwt_counter))
                    p_peaks.append(temp_dwt)
                elif i == 3:
                    print("Annotate T-peaks. If previous peak was the last T-peaks type ok")
                    temp_dwt = input("Annotate {} T-peaks: ".format(dwt_counter))
                    t_peaks.append(temp_dwt)
                elif i == 4:
                    print("Annotate T-offsets. If previous peak was the last T-offsets type ok")
                    temp_dwt = input("Annotate {} T-offsets: ".format(dwt_counter))
                    t_offsets.append(temp_dwt)
                elif i == 5:
                    print("Annotate T-onsets. If previous peak was the last T-onsets type ok")
                    temp_dwt = input("Annotate {} T-onsets: ".format(dwt_counter))
                    t_onsets.append(temp_dwt)
                else:
                    print("This is an unknown state")
        p_offsets = np.asarray(p_offsets)
        p_onsets = np.asarray(p_onsets)
        p_peaks = np.asarray(p_peaks)
        t_peaks = np.asarray(t_peaks)
        t_offsets = np.asarray(t_offsets)
        t_onsets = np.asarray(t_onsets)
        
    try:
        _, waves_peak = nk.ecg_delineate(ecg_data[1], rpeaks, sampling_rate=sampling_rate)
    except:
        print("Could not find S and Q waves")
        fig = go.Figure()
        fig.add_trace(go.Scatter(x=signal_length, y=ecgdata[1],
            mode='lines',
            name='lines'))
        fig.show()
        print("Please help annotate")
        Q_peaks = []
        S_peaks = []
        for i in range(2):
            temp_peak = 0
            peak_counter = 0
            while temp_peak != "ok":
                peak_counter = peak_counter + 1 
                if i == 0:
                    print("Annotate Q-peaks. If previous peak was the last P-offset type ok")
                    temp_peak = input("Annotate {} Q-peaks: ".format(peak_counter))
                    Q_peaks.append(temp_peak)
                elif i == 1:
                    print("Annotate P-onsets. If previous peak was the last P-onsets type ok")
                    temp_peak = input("Annotate {} P-onsets: ".format(peak_counter))
                    S_peaks.append(temp_peak)
                else:
                    print("This is an unknown state")
        Q_peaks = np.asarray(Q_peaks)
        S_peaks = np.asarray(S_peaks)
    manual_annot[:,n_annot] = rpeaks, p_offsets, p_onsets, p_peaks, t_peaks, t_offsets, t_onsets, Q_peaks, S_peaks
```
